## Most Successful Nation in Olympic History

```python
#sql medal overall count
medals1=pysqldf('select Year, NOC_ID, Sport_ID, count(distinct Medal_ID) as medals_won, count(distinct case when Medal_ID=1 then 1 else null end) as gold, count(distinct case when Medal_ID=2 then 1 else null end) as silver, count(distinct case when Medal_ID=3 then 1 else null end) as bronze from results where Medal_ID !=4 group by Year, NOC_ID, Sport_ID order by Year;')
```

```python
all_time_medals_ov=pysqldf('select nationality.NOC as Nations, sum(medals_won) as all_time_medals, sum(gold) as gold_medals, sum(silver) as silver_medals, sum(bronze) as bronze_medals from medals1 left join nationality on medals1.NOC_ID=nationality.NOC_ID group by nationality.NOC order by sum(medals_won) desc limit 10')
all_time_medals_ov.set_index('Nations', inplace=True)
```

SQL Syntax
*/Getting the overall medal count for the different nations/*

select

	nationality.NOC as Nations,
	sum(medals_won) as all_time_medals,
	sum(gold) as gold_medals,
	sum(silver) as silver_medals,
	sum(bronze) as bronze_medals

from(

	select
		Year,
		NOC_ID,
		Sport_ID,
		count(distinct Medal_ID) as medals_won,
		count(distinct case when Medal_ID=1 then 1 else null end) as gold,
		count(distinct case when Medal_ID=2 then 1 else null end) as silver,
		count(distinct case when Medal_ID=3 then 1 else null end) as bronze
		from results
		where Medal_ID !=4
		group by Year, NOC_ID, Sport_ID order by Year
        
) as medals1

	left join nationality
    
		on medals1.NOC_ID=nationality.NOC_ID
        
group by nationality.NOC

order by sum(medals_won) desc limit 10;

## From 1896-2016 Number of Events for Women and Men

### Hypothesis=> In 2016 women and men have an equal number of particpants

```python
#sql to get the numbers of sports events of men and women
men_women=pysqldf('Select results.Year, olympic_games.Games, count(distinct case when results.Sex_ID=1 then results.Sport_ID else null end) as male_medal_count, count(distinct case when results.Sex_ID=2 then results.Sport_ID else null end) as female_medal_count from results left join olympic_games on results.Games_ID=olympic_games.Games_ID group by results.Year, olympic_games.Games')
```

```python
#sql difference between men and women events (in numbers)
dif_men_women=pysqldf('select Year, Games, sum(male_medal_count) as events_men, sum(female_medal_count) as events_women, (female_medal_count - male_medal_count) as difference_between_men_women from men_women group by Year,Games;')
```

SQL Syntax
*/Difference in Numbers between Male and Female/*

Select

    Year,
	Games,
	sum(male_medal_count) as events_men,
	sum(female_medal_count) as events_women,
	(sum(female_medal_count) - sum(male_medal_count)) as difference_between_men_women
    
from (

	Select
		results.Year,
		olympic_games.Games,
		count(distinct case when results.Sex_ID=1 then results.Sport_ID else null end) as male_medal_count,
		count(distinct case when results.Sex_ID=2 then results.Sport_ID else null end) as female_medal_count
	from results
	left join olympic_games
		on results.Games_ID=olympic_games.Games_ID
	group by results.Year, olympic_games.Games
) as men_women

group by Year, Games;

```python
dif_men_women.set_index('Year', inplace=True)
```

```python
dif_men_women[['events_men', 'events_women', 'difference_between_men_women']].plot.line(
    color=['blue', 'red', 'green'])
plt.legend(loc='lower right')
```

```python
dif_men_women[['events_men', 'events_women', 'difference_between_men_women']].plot.line(
    color=['blue', 'red', 'green'], legend=None)
plt.title('Differencen in number of events between Men and Women in 120 years', size=12, fontweight='bold')
```

As we can see, men generally had more events than women, but the trend shows that the difference between men and women is getting smaller.

We have to reject the hypothesis because women did not have the same number of events as men in 2016.
