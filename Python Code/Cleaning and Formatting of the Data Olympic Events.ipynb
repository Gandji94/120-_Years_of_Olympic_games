### Medal

```python
sports_events
```

```python
#creating a new dataframe with the different medal types
medal_df=pd.DataFrame(sports_events.Medal.unique().tolist(), columns=['Medal'])
```

```python
#Nan is considered as not winning a medal, I have transformed it to 'No_Medal'
medal_df=medal_df.fillna('No_Medal')
```

```python
medal_df
```

```python
#using lambda to allocate a number to each medal type
medal_df['Medal_ID']=medal_df['Medal'].apply(lambda x: 1 if x=='Gold' else (2 if x=='Silver' else
                                                                            (3 if x=='Bronze' else 4)))
```

```python
medal_df
```

```python
#sorting values
medal_df.sort_values('Medal_ID', inplace=True)
```

```python
#setting a new index
medal_df.set_index('Medal_ID', inplace=True)
```

```python
#saving it on a csv file
medal_df.to_csv('/Users/alexandergandji/Desktop/Course Folder/Capstone Project SQL Data Science /Sports/olympic_medal.csv')
```

### olympic games

```python
sports_events.head()
```

```python
#using existing dataframe and formate it so no duplicates remain and every olympic game appears once
games=sports_events[['Games', 'Year', 'Season', 'City']].drop_duplicates()
```

```python
#sorting values based on Games
games.sort_values('Games', inplace=True)
```

```python
#using lambda to allocate to summer and winter games an unique number/ID
games['Season_ID']=games['Season'].apply(lambda x: 1 if x=='Summer' else 2)
```

```python
len(games.Games)
```

```python
#allocating every game an unique number/ID
games['Games_ID']=range(1,53)
```

```python
#droping column 'Season', this could have also been done with reindex()
games.drop('Season',1, inplace=True)
```

```python
games.head()
```

```python
#reindexing the columns
games_df=games.reindex(columns=['Games', 'Games_ID', 'Year', 'Season_ID', 'City'])
```

```python
#setting a new index
games_df.set_index('Games_ID', inplace=True)
```

```python
#saving it on a csv file
games_df.to_csv('/Users/alexandergandji/Desktop/Course Folder/Capstone Project SQL Data Science /Sports/olympic_games.csv')
```

### athletes

```python
sports_events.head()
```

```python
#getting an unique list of athletes
athlete=sports_events.Name.unique().tolist()
```

```python
#creating a new dataframe
athlete=pd.DataFrame(athlete, columns=['Name'])
```

```python
#sorting dataframe by name in ascending order
athlete.sort_values('Name', inplace=True)
```

```python
#removing the extra space
athlete.replace('  Gabrielle Marie "Gabby" Adcock (White-)', 'Gabrielle Marie "Gabby" Adcock (White-)',
                inplace=True)
athlete.replace(' Eleonora Margarida Josephina Scmitt', 'Eleonora Margarida Josephina Scmitt', inplace=True)
athlete.replace(' Jean Hauptmanns', 'Jean Hauptmanns', inplace=True)
athlete.replace(' Luis ngel Fernando de los Santos Grossi', 'Luis ngel Fernando de los Santos Grossi', inplace=True)
athlete.replace(' Th Anh', 'Th Anh', inplace=True)
athlete.replace(' Th Ngn Thng', 'Th Ngn Thng', inplace=True)
athlete.replace(' Tin Tun', 'Tin Tun', inplace=True)
athlete.sort_values('Name', inplace=True)
```

```python
#allocating an unique number/ID to every athlete and merging the dataframe with the orginal data to get the gender
athlete['Athlete_ID']=range(1,134788)
athlete=athlete.merge(sports_events[['Name', 'Sex']], how='left').drop_duplicates()
```

```python
#using lambda to allocate an unique number to every gender
athlete['Sex_ID']=athlete.Sex.apply(lambda x: 1 if x=='M' else 2)
```

```python
#setting a new index
athlete.set_index('Athlete_ID', inplace=True)
```

```python
ath1=athlete.dropna()
```

```python
#saving it on a csv file
ath1.to_csv('/Users/alexandergandji/Desktop/Course Folder/Capstone Project SQL Data Science /Sports/athlete.csv')
```

### results

```python
#merging orginal data with the nations dataframe
results1=sports_events.merge(nat_df1.reset_index(), how='left')
```

```python
#adding a new column, which contains the medal id; using lambda
results1['Medal_ID']=results1.Medal.apply(lambda x: 1 if x=='Gold' else
                                          (2 if x=='Silver' else (3 if x=='Bronze' else 4)))
```

```python
#removing extra space from athletes name
results1.replace('  Gabrielle Marie "Gabby" Adcock (White-)', 'Gabrielle Marie "Gabby" Adcock (White-)',
                 inplace=True)
results1.replace(' Eleonora Margarida Josephina Scmitt', 'Eleonora Margarida Josephina Scmitt', inplace=True)
results1.replace(' Jean Hauptmanns', 'Jean Hauptmanns', inplace=True)
results1.replace(' Luis ngel Fernando de los Santos Grossi', 'Luis ngel Fernando de los Santos Grossi',
                 inplace=True)
results1.replace(' Th Anh', 'Th Anh', inplace=True)
results1.replace(' Th Ngn Thng', 'Th Ngn Thng', inplace=True)
results1.replace(' Tin Tun', 'Tin Tun', inplace=True)
```

```python
#merging results dataframe with the games dataframe
results2=results1.merge(games_df.reset_index(), how='left')
```

```python
#merging results dataframe with the athlete dataframe
results3=results2.merge(athlete.reset_index(), how='left')
```

```python
#dropping 'Sport' column=> this column just gave an general overview of the events
results3.drop('Sport',1,inplace=True)
```

```python
#renaming Event to Sport=> the new Sport column has a more detailed overview of the different events
results3.rename(columns={'Event':'Sport'}, inplace=True)
```

```python
#merging results dataframe with sports dataframe 
results4=results3.merge(sports.reset_index(), how='left')
```

```python
#sort result dataframe by Year
results4.sort_values('Year', inplace=True)
```

```python
results4
```

```python
#due to the fact that we have data points which did not have any value in the Age, Height and Weight column we have
#used ffill()
results4['Age']=results4.Age.ffill(axis=0)
#applying ffill() in the first row of the Age column
results4.Age.replace(results4.Age.iloc[0],21, inplace=True)
#transforming Age from float to integer
results4['Age']=results4.Age.apply(np.int64)
```

```python
results4=results4.copy()
```

```python
#applying ffill() in the first row of the Height column
results4['Height']=results4.Height.replace(results4.Height.iloc[0],176)
#transforming Height from float to integer
results4['Height']=results4.Height.apply(np.int64)

### 176 was chosen, because the height was found at other athletes in the same time period and disceplines
```
